```python
import os
import openpyxl
import xlrd
import regex
from openpyxl import workbook
```

```python
wb_sheet=openpyxl.load_workbook('Bowler Pattern.xlsx')
wb_matches=openpyxl.load_workbook('all_matches.xlsx')
```

```python
ws_sheet=wb_sheet.active
ws_matches=wb_matches.active
```

```python
row_idx=1
innings=0
for rowx in range(193500,ws_matches.max_row+1):
    year_val = ws_matches.cell(row=rowx,column=2).value
    year = int(str(year_val)[0:4])
    if  year == 2021 and innings <3:
        
        #when innings changes, row change
        if innings!= ws_matches.cell(row=rowx,column=5).value:
            row_idx+=1
            innings=ws_matches.cell(row=rowx,column=5).value
            #match_id - 1,innings - 2, team name - 3
            match_id=ws_matches.cell(row=rowx,column=1).value
            team_name=ws_matches.cell(row=rowx,column=8).value
            
            ws_sheet.cell(row=row_idx,column=1,value=match_id)
            ws_sheet.cell(row=row_idx,column=2,value=innings)
            ws_sheet.cell(row=row_idx,column=3,value=team_name)
            
        #when over changes
        over=float(ws_matches.cell(row=rowx,column=6).value)
        k = int(over)
        if round(over-k,1) == 0.1 and k<6:
            #add to over column
            bowler=ws_matches.cell(row=rowx,column=11).value
            if k==0:
                ws_sheet.cell(row=row_idx,column=4,value=bowler)
            if k==1:
                ws_sheet.cell(row=row_idx,column=5,value=bowler)
            if k==2:
                ws_sheet.cell(row=row_idx,column=6,value=bowler)
            if k==3:
                ws_sheet.cell(row=row_idx,column=7,value=bowler)
            if k==4:
                ws_sheet.cell(row=row_idx,column=8,value=bowler)
            if k==5:
                ws_sheet.cell(row=row_idx,column=9,value=bowler)
                
            
```

```python
wb_sheet.save("Bowler Pattern.xlsx")
```
